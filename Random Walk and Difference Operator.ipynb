```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 10000
x = [0]
mu = 0
sigma = 1
for i in range(n):
    x.append(x[len(x)-1]+np.random.normal(mu, sigma, 1)[0])
```

```python
plt.plot(x)
```

```python
diff_x = np.diff(x)
```

```python
plt.plot(diff_x)
```

```python
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(x,lags=50);
```

```python
plot_acf(diff_x,lags=50);
```
